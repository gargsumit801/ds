# Exploratory Data Analysis (EDA) for Super Store US

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("US_Superstore_data1.csv")
```

```python
df.info()
```

```python
df['Order Date'] = df['Order Date'].astype('datetime64[ns]')
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Order Date'].info()
```

```python
df['Order Date'].value_counts()
```

### Univariate analysis

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Segment',data=df)
plt.title('Count of Segment')
plt.savefig('Count of Segment')
plt.show()
```

```python
df['Ship Mode'].value_counts()
```

```python
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(8,5))
plt.pie(y,labels=x,autopct = '%0.2f%%',startangle=70)
plt.legend(loc=2)
plt.show()
```

```python
## Bivariate analysis

plt.figure(figsize=(6,3))
sns.countplot(x='Category',data=df,hue='Segment')
plt.title('category wise segment')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Category',data=df)
plt.title('Count of Category')
plt.show()
```

```python
df['Category'].unique()
```

```python
# bivariate analysis

plt.figure(figsize=(8,3))
sns.countplot(x='Category',data=df[df['Category']=='Furniture'] ,hue='Sub-Category')
plt.show()
```

```python
# bivariate analysis

plt.figure(figsize=(8,4))
sns.countplot(x='Category',data=df[df['Category']=='Office Supplies'],hue='Sub-Category')
plt.show()
```

```python
# bivariate analysis

plt.figure(figsize=(8,4))
sns.countplot(x='Category',data=df[df['Category']=='Technology'],hue='Sub-Category')
plt.show()
```
